```python
import tkinter as tk
from tkinter import *
import mysql.connector
from cryptography.fernet import Fernet
def goto_encryption():
# Database connection configuration
   db_config = {
      "host": "localhost",#hostname
      "user": "billgates",#user_name
      "password": "pwdbgates",#password
      "database": "project",#database_name
    }

# Create a tkinter window
   window = tk.Tk()
   window.title("Data Encryption and Decryption")

# Create a Text widget to display data
   text_widget = tk.Text(window, height=30, width=50)
   text_widget.pack()

# Generate an AES key (you should store and retrieve this key securely in a real application)
   key = Fernet.generate_key()
   cipher_suite = Fernet(key)

# Function to retrieve data from the MySQL database
   def retrieve_data():
    try:
      connection = mysql.connector.connect(**db_config)
      cursor = connection.cursor()
      cursor.execute("SELECT * FROM department")
      data = cursor.fetchall()
      connection.close()
      return data

    except mysql.connector.Error as err:
      print("Error:", err)
      return []

# Function to encrypt data using AES
   def encrypt_data():
      data = text_widget.get("1.0", "end-1c")
      encrypted_data = cipher_suite.encrypt(data.encode()).decode()
      text_widget.delete("1.0", "end")
      text_widget.insert("1.0", encrypted_data)

# Function to decrypt data using AES
   def decrypt_data():
      encrypted_data = text_widget.get("1.0", "end-1c")
      decrypted_data = cipher_suite.decrypt(encrypted_data.encode()).decode()
      text_widget.delete("1.0", "end")
      text_widget.insert("1.0", decrypted_data)

# Create buttons for actions
   retrieve_button = tk.Button(window, text="Retrieve Data from Database",bg="lightgray",font=("Trajan pro",25), command=lambda: text_widget.insert("1.0", "\n".join([str(row) for row in retrieve_data()])))

   encrypt_button = tk.Button(window, text="Encrypt",bg="lightgray",font=("Trajan pro",25), command=encrypt_data)
   decrypt_button = tk.Button(window, text="Decrypt",bg="lightgray",font=("Trajan pro",25), command=decrypt_data)

   retrieve_button.pack(pady=2)
   encrypt_button.pack(pady=5)
   decrypt_button.pack(pady=5)

root = Tk()
ll=Label(text="RBVRR WOMEN'S COLLEGE",font=("Trajan Pro",40,"bold","underline","italic"),foreground="black")
l1=Label(text="MINOR PROJECT",font=("Trajan Pro",30,"italic","bold"),foreground="black",pady="20")
l15=Label(text="ON",font=("Trajan Pro",30,"italic","bold"),foreground="black",pady="15")

l2=Label(text="Secure Data Encryption System with AES Algorithm",font=("Trajan Pro",30,"bold","underline","italic"),foreground="black",pady="20")
ll2=Label(text="Course title:Network Security ",font=("Trajan Pro",25,"bold","underline","italic"),foreground="black",pady="20")

l3=Label(text="BY:",font=("times new roman",20,"bold","italic"),foreground="black",pady="5")

l6=Label(text="S.Alekhya-117222504023",font=("Trajan Pro",20,"bold"),foreground="black",pady="5")


ll.pack()
l1.pack()
l15.pack()

l2.pack()

ll2.pack()

l3.pack()
l6.pack()
	

label = Label(root, text="Want to perform Encryption and Decryption?", font=("Times new roman", 25,"bold"))
label.pack(pady=20)

button_yes = Button(root, text="Yes",bg="green", font=("Arial", 20), command=goto_encryption)
button_yes.pack(pady=5)
button_no = Button(root, text="No",bg="red" ,font=("Arial", 25))
button_no.pack(pady=5)

root.mainloop()


# Start the tkinter main loop
window.mainloop()
```
